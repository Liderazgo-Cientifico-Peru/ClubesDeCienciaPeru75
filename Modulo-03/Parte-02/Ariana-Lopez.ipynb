```python
MUJERES_NO_INGRESARON = 8962
MUJERES_INGRESARON = 1135
MUJERES_TOTAL = MUJERES_NO_INGRESARON + MUJERES_INGRESARON

HOMBRES_NO_INGRESARON = 26573
HOMBRES_INGRESARON = 5846
HOMBRES_TOTAL = HOMBRES_NO_INGRESARON + HOMBRES_INGRESARON

NO_INGRESARON = MUJERES_NO_INGRESARON + HOMBRES_NO_INGRESARON
INGRESARON = MUJERES_INGRESARON + HOMBRES_INGRESARON
TOTAL = MUJERES_TOTAL + HOMBRES_TOTAL


PORCENTAJE_MUJERES_TOTAL = (MUJERES_TOTAL / TOTAL)*100
print(f"Porcentaje de mujeres que postularon: {PORCENTAJE_MUJERES_TOTAL} %")

PORCENTAJE_MUJERES_INGRESARON = (MUJERES_INGRESARON / INGRESARON) * 100
print(f"Porcentaje de mujeres que postularon e ingresaron: {PORCENTAJE_MUJERES_INGRESARON} %")

PORCENTAJE_MUJERES_NO_INGRESARON = (MUJERES_NO_INGRESARON / NO_INGRESARON) * 100
print(f"Porcentaje de mujeres que postularon y no ingresaron: {PORCENTAJE_MUJERES_INGRESARON} %")
```

```python
MUJERES_UNI = df[df['SEXO']=='FEMENINO']['INGRESO'].value_counts()
HOMBRES_UNI = df[df['SEXO']=='MASCULINO']['INGRESO'].value_counts()
```

```python

sns.barplot(x=HOMBRES_UNI.index, y=HOMBRES_UNI.values , label = 'Masculino', color='blue')
sns.barplot(x=MUJERES_UNI.index, y=MUJERES_UNI.values , label = 'Femenino', color='pink')
plt.ylabel('CANTIDAD POR GÉNERO')
plt.xlabel('POSTULANTES A LA UNI SEGÚN EL GÉNERO DURANTE 4 AÑOS')
plt.show()
```
