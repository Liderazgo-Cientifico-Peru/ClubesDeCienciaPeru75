```python
def categorize_origin(province):
    """Categoriza una provincia como 'Zona Urbana' o 'Zona Rural'."""
    # Basado en la exploración anterior, 'LIMA' y 'PROV CONST DEL CALLAO' se consideran urbanas.
    # Todas las demás provincias se consideran rurales.
    if province in ["LIMA", "PROV CONST DEL CALLAO", "PROV. CONST. DEL CALLAO"]:
        return "Zona Urbana"
    else:
        return "Zona Rural"

df['Origin_Category'] = df['COLEGIO_PROV'].apply(categorize_origin)
display(df[['COLEGIO_PROV', 'Origin_Category']].head())
```

```python
# Calcular el recuento de estudiantes admitidos y no admitidos por categoría de origen
admission_status_counts = df.groupby(['Origin_Category', 'INGRESO']).size().reset_index(name='count')
display("Recuento de estado de admisión por categoría de origen:")
display(admission_status_counts)

# Calcular el recuento total para cada categoría de origen
total_counts = df.groupby('Origin_Category').size().reset_index(name='total')

# Fusionar el recuento de estado de admisión con el recuento total para calcular porcentajes
admission_status_counts = pd.merge(admission_status_counts, total_counts, on='Origin_Category')

# Calcular el porcentaje de volumen para cada categoría
admission_status_counts['percentage_volume'] = (admission_status_counts['count'] / admission_status_counts['total']) * 100

# Calcular el porcentaje de admisión dentro de cada categoría
admission_status_counts['percentage_admission'] = 0.0
admission_status_counts.loc[admission_status_counts['INGRESO'] == 'SI', 'percentage_admission'] = (
    admission_status_counts[admission_status_counts['INGRESO'] == 'SI']['count'] /
    admission_status_counts[admission_status_counts['INGRESO'] == 'SI']['total']
) * 100


display("\nRecuento y porcentajes de estado de admisión por categoría de origen:")
display(admission_status_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de barras comparando estudiantes admitidos y no admitidos por categoría de origen
plt.figure(figsize=(10, 6))
sns.barplot(x='Origin_Category', y='count', hue='INGRESO', data=admission_status_counts, palette='viridis')
plt.title("Número de estudiantes por categoría de origen y estado de admisión")
plt.xlabel("Categoría de origen")
plt.ylabel("Número de estudiantes")
plt.show()

# Gráfico de barras comparando el porcentaje de volumen por categoría de origen y estado de admisión
plt.figure(figsize=(10, 6))
sns.barplot(x='Origin_Category', y='percentage_volume', hue='INGRESO', data=admission_status_counts, palette='viridis')
plt.title("Porcentaje del total de estudiantes por categoría de origen y estado de admisión")
plt.xlabel("Categoría de origen")
plt.ylabel("Porcentaje del total de estudiantes")
plt.show()

# Gráfico de barras comparando el porcentaje de admisión dentro de cada categoría de origen
plt.figure(figsize=(10, 6))
sns.barplot(x='Origin_Category', y='percentage_admission', data=admission_status_counts[admission_status_counts['INGRESO'] == 'SI'], palette='viridis')
plt.title("Porcentaje de admisión dentro de cada categoría de origen")
plt.xlabel("Categoría de origen")
plt.ylabel("Porcentaje de admisión")
plt.show()
```

```python
def categorize_geographical_region(row):
    """Categoriza la región geográfica de un estudiante basado en el departamento y provincia del colegio."""
    depa = row['COLEGIO_DEPA']
    prov = row['COLEGIO_PROV']

    lima_metropolitana_provinces = {
        'LIMA': ['LIMA', 'BARRANCA', 'CAJATAMBO', 'CANTA', 'CAÑETE', 'HUAURA', 'HUARAL', 'HUAROCHIRI', 'OYON', 'YAUYOS'],
        'CALLAO': ['PROV CONST DEL CALLAO', 'PROV. CONST. DEL CALLAO']
    }

    costa_depas = ['TUMBES', 'PIURA', 'LAMBAYEQUE', 'LA LIBERTAD', 'ÁNCASH', 'ANCASH', 'LIMA', 'ICA', 'AREQUIPA', 'MOQUEGUA', 'TACNA', 'CALLAO']
    sierra_depas = ['CAJAMARCA', 'HUANUCO', 'HUÁNUCO', 'PASCO', 'JUNIN', 'JUNÍN', 'HUANCAVELICA', 'AYACUCHO', 'APURIMAC', 'APURÍMAC', 'CUSCO', 'PUNO']
    selva_depas = ['AMAZONAS', 'SAN MARTIN', 'SAN MARTÍN', 'LORETO', 'UCAYALI', 'MADRE DE DIOS']

    # Specific categorization for Lima Metropolitana sub-regions
    if depa == 'LIMA':
        if prov in ['LIMA', 'CAÑETE', 'BARRANCA', 'HUAURA', 'HUARAL', 'HUAROCHIRI', 'YAUYOS']:
            # This simplified approach doesn't distinguish between Lima Norte, Sur, and Centro based on province alone
            # A more detailed mapping of districts to sub-regions would be needed for that.
            return 'Lima Metropolitana - Lima'
        elif prov in ['CANTA', 'CAJATAMBO', 'OYON']:
             # This simplified approach doesn't distinguish between Lima Norte, Sur, and Centro based on province alone
            # A more detailed mapping of districts to sub-regions would be needed for that.
            return 'Lima Metropolitana - Lima'
        else:
             # Handle other provinces in Lima department not typically considered Lima Metropolitana
             return 'Sierra' if prov in ['HUAROCHIRI'] else 'Costa'


    elif depa == 'CALLAO':
        return 'Lima Metropolitana - Callao'

    # Categorization for other regions
    if depa in costa_depas:
        return 'Costa'
    elif depa in sierra_depas:
        return 'Sierra'
    elif depa in selva_depas:
        return 'Selva'
    else:
        return 'Otro/Desconocido'

df['Geographical_Region'] = df.apply(categorize_geographical_region, axis=1)

display(df[['COLEGIO_DEPA', 'COLEGIO_PROV', 'Geographical_Region']].head())
display(df['Geographical_Region'].value_counts())
```

```python
admitted_students_df = df[df['INGRESO'] == 'SI'].copy()

# Calculate the count of admitted students by gender
admission_counts_gender = admitted_students_df.groupby('SEXO').size().reset_index(name='count')
display("Recuento de estudiantes admitidos por género:")
display(admission_counts_gender)

# Calculate the total number of students by gender (admitted and non-admitted combined)
total_counts_gender = df.groupby('SEXO').size().reset_index(name='total')
display("\nRecuento total de estudiantes por género:")
display(total_counts_gender)

# Merge the admitted counts with total counts
gender_admission_metrics = pd.merge(admission_counts_gender, total_counts_gender, on='SEXO')

# Calculate the percentage of admitted students for each gender relative to the total number of admitted students
total_admitted = admitted_students_df.shape[0]
gender_admission_metrics['percentage_of_admitted_total'] = (gender_admission_metrics['count'] / total_admitted) * 100

# Calculate the percentage of admitted students within each gender category (admission rate)
gender_admission_metrics['admission_rate'] = (gender_admission_metrics['count'] / gender_admission_metrics['total']) * 100

display("\nMétricas de admisión por género:")
display(gender_admission_metrics)
```

```python
print("Valores únicos en MODALIDAD:")
print(df['MODALIDAD'].unique())
print("\nRecuento de valores en MODALIDAD:")
print(df['MODALIDAD'].value_counts())
```

```python
admitted_students_df = df[df['INGRESO'] == 'SI'].copy()

# Calculate the count of admitted students by modality
admission_counts_modality = admitted_students_df.groupby('MODALIDAD').size().reset_index(name='count')

# Calculate the total number of students by modality (admitted and non-admitted combined)
total_counts_modality = df.groupby('MODALIDAD').size().reset_index(name='total')

# Merge the admitted counts with total counts
modality_admission_metrics = pd.merge(admission_counts_modality, total_counts_modality, on='MODALIDAD')

# Calculate the percentage of admitted students for each modality relative to the total number of admitted students
total_admitted = admitted_students_df.shape[0]
modality_admission_metrics['percentage_of_admitted_total'] = (modality_admission_metrics['count'] / total_admitted) * 100

# Calculate the percentage of admitted students within each modality category (admission rate)
modality_admission_metrics['admission_rate'] = (modality_admission_metrics['count'] / modality_admission_metrics['total']) * 100

display("Métricas de admisión por modalidad:")
display(modality_admission_metrics)
```

```python
admitted_students_df = df[df['INGRESO'] == 'SI'].copy()

# Calculate the count of admitted students by geographical region
admission_counts_region = admitted_students_df.groupby('Geographical_Region').size().reset_index(name='count')

# Calculate the total number of students by geographical region (admitted and non-admitted combined)
total_counts_region = df.groupby('Geographical_Region').size().reset_index(name='total')

# Merge the admitted counts with total counts
region_admission_metrics = pd.merge(admission_counts_region, total_counts_region, on='Geographical_Region')

# Calculate the percentage of admitted students for each geographical region relative to the total number of admitted students
total_admitted = admitted_students_df.shape[0]
region_admission_metrics['percentage_of_admitted_total'] = (region_admission_metrics['count'] / total_admitted) * 100

# Calculate the percentage of admitted students within each geographical region (admission rate)
region_admission_metrics['admission_rate'] = (region_admission_metrics['count'] / region_admission_metrics['total']) * 100

display("Métricas de admisión por región geográfica:")
display(region_admission_metrics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de barras: Número de estudiantes por región geográfica y estado de admisión
plt.figure(figsize=(12, 7))
sns.barplot(x='Geographical_Region', y='count', hue='INGRESO', data=df.groupby(['Geographical_Region', 'INGRESO']).size().reset_index(name='count'), palette='viridis')
plt.title("Número de estudiantes por región geográfica y estado de admisión")
plt.xlabel("Región Geográfica")
plt.ylabel("Número de estudiantes")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Gráfico de barras: Tasa de admisión por región geográfica
plt.figure(figsize=(12, 7))
sns.barplot(x='Geographical_Region', y='admission_rate', data=region_admission_metrics, palette='viridis')
plt.title("Tasa de admisión por región geográfica")
plt.xlabel("Región Geográfica")
plt.ylabel("Tasa de Admisión (%)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Gráfico de barras: Número de estudiantes por género y estado de admisión
plt.figure(figsize=(8, 6))
sns.barplot(x='SEXO', y='count', hue='INGRESO', data=df.groupby(['SEXO', 'INGRESO']).size().reset_index(name='count'), palette='viridis')
plt.title("Número de estudiantes por género y estado de admisión")
plt.xlabel("Género")
plt.ylabel("Número de estudiantes")
plt.show()

# Gráfico de barras: Tasa de admisión por género
plt.figure(figsize=(8, 6))
sns.barplot(x='SEXO', y='admission_rate', data=gender_admission_metrics, palette='viridis')
plt.title("Tasa de admisión por género")
plt.xlabel("Género")
plt.ylabel("Tasa de Admisión (%)")
plt.show()

# Gráfico de barras: Número de estudiantes admitidos por modalidad
plt.figure(figsize=(14, 8))
sns.barplot(x='MODALIDAD', y='count', data=modality_admission_metrics, palette='viridis')
plt.title("Número de estudiantes admitidos por modalidad")
plt.xlabel("Modalidad de Admisión")
plt.ylabel("Número de estudiantes admitidos")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()

# Gráfico de barras: Tasa de admisión por modalidad
plt.figure(figsize=(14, 8))
sns.barplot(x='MODALIDAD', y='admission_rate', data=modality_admission_metrics, palette='viridis')
plt.title("Tasa de admisión por modalidad")
plt.xlabel("Modalidad de Admisión")
plt.ylabel("Tasa de Admisión (%)")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

## Resumen:

### Descubrimientos clave al analizar la data

*   **Región Geográfica:**
    *   La mayoria de los estudiantes admitidos (80.49%) son de 'Lima Metropolitana - Lima'.
    *   La taza de admisión varía significativamente por regiones, con 'Otro/Desconocido' teniendo mayor taza (38.16%) y 'Selva' la menor (10.28%).
    *   'Lima Metropolitana - Callao', 'Costa', y  'Sierra' tienen tazas de admisión similar, al rededor de 14-15%.
*   **Género:**
    *   Un gran número de varones han sido admitidos (5846) a comparación de estudiantes femeninas (1135).
    *   Estudiantes masculinos representan el 83.74% del total de estudiantes admitidos, mientras que las estudiantes muujeres representan el 16.26%.
    *   La taza de admisión es mayor para estudiantes hombres (18.03%) que para estudantes femeninos (11.24%).
*   **Modalidad de admisión:**
    *   La modalidad 'ORDINARIO' representa el mayor número de estudiantes admitidos (4070).
    *   Muchas modalidades extraordinarias, como la de 'EXTRAORDINARIO 2 – INGRESO DIRECTO CEPRE' (1204 admitidos) y 'EXTRAORDINARIO INGRESO DIRECTO CEPRE-UNI' (562 admitidos), también contribuye significativamente al pozo de admitidos.
    *   Las tazas de admisión varían considerablemente entre modalidades, con algunas modalidades extraordinarias teniendo el 100% de taza de admisión (sin embargo con un número menor de postulantes).
    *   Modalidades con un gran volumen ('ORDINARIO', 'INGRESO ESCOLAR NACIONAL') tienden a tener bajas tazas de admisión comparado con algunas modalidades extraordinarias con bajo volumen.
