Definamos zonas urbanas en nuestro dataset

```python
DEPARTAMENTOS_URBANOS = ['CALLAO', 'LIMA', 'TUMBES', 'ICA', 'AREQUIPA', 'TACNA', 'LAMBAYEQUE']
```

```python
df['ZONA'] = df['DOMICILIO_DEPA'].apply(lambda x: 'URBANO' if x in DEPARTAMENTOS_URBANOS else 'RURAL')
```

```python
df.sample(8)
```

```python
df['ZONA'].value_counts()
```

```python
ADMITIDOS_POR_ZONA = df[df['INGRESO']=='SI']['ZONA'].value_counts()
NO_ADMITIDOS_POR_ZONA = df[df['INGRESO']=='NO']['ZONA'].value_counts()
```

Nuestro primer histograma

```python
sns.barplot(x=NO_ADMITIDOS_POR_ZONA.index, y=NO_ADMITIDOS_POR_ZONA.values , label = 'No ingresaron', color='red')
sns.barplot(x=ADMITIDOS_POR_ZONA.index, y=ADMITIDOS_POR_ZONA.values , label = 'Ingresaron', color='blue')
plt.ylabel('Numero de postulantes')
plt.show()
```
